```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.show()
```

```python
x = np.linspace(-5, 10, 500)
a = -40
b= 0
c= 1000
d= 5
plt.plot(x, d*(x**4) + a*(x**3) + b*(x**2) + c*x)
plt.title('A simple chirp')
plt.show()
```

```python
x = np.arange(0., 10., 0.2)
a = 1
b= 10
c= 10
d= -.1
plt.plot(x, d*(x**4), 'r--',  a*(x**3), 'bs',  b*(x**2), 'g^', c*x , 'r^')
plt.show()
```
